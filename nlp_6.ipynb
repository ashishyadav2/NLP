## Stop words


```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS # to get stop words present in english lang
```

```python
nlp = spacy.load('en_core_web_sm')
doc = nlp("we just opened our wings, the flying part is coming soon")
```

```python
for token in doc:
  if token.is_stop: # to check if token is a stop word
    print(token)
```

```python
removed_stop_words = [token.text for token in doc if not token.is_stop and not token.is_punct]
removed_stop_words
```

```python
def remove_stop_words(col):
  for i in range(len(df[col])):
    doc = nlp(df[col][i])
    df[col][i] = ' '.join([token.text for token in doc if not token.is_stop and not token.is_punct])
```
