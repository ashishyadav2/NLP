```python
import spacy
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
doc = nlp("dog cat banana kem")
# it will print if there is a vector present in large model, OOV is True if the word is out of vocabulary
for token in doc:
  print(f"{token.text}, Vector: {token.has_vector}, OOV:{token.is_oov}")
```

```python
# to get vector shape
# every word will have vector size 300
base_token = nlp("bread")
base_token.vector.shape
```

```python
# it will tell the word to which it most similar based on their vectors
# similarity depends upon the context they have been appeared while training the large model of spacy
doc = nlp("bread sandwich burger car tiger human wheat")
for token in doc:
  print(f"{base_token}<->{token.text}: {token.similarity(base_token)}")
```

```python
king = nlp.vocab["king"].vector
man = nlp.vocab["man"].vector
woman = nlp.vocab["woman"].vector
queen = nlp.vocab["queen"].vector
result = king - man + woman
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity([result],[queen])
```
